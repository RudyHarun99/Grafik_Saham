```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
excl = pd.read_csv('EXCL.JK.CSV',parse_dates=['Date'],index_col='Date')
fren = pd.read_csv('FREN.JK.CSV',parse_dates=['Date'],index_col='Date')
isat = pd.read_csv('ISAT.JK.CSV',parse_dates=['Date'],index_col='Date')
tlkm = pd.read_csv('TLKM.JK.CSV',parse_dates=['Date'],index_col='Date')
```

# EXCL

```python
excl = excl.sort_index()
excl = excl.resample('D').sum()
excl.iloc[:,:5] = excl.iloc[:,:5].replace(0, np.NaN)
excl = excl.fillna(method='ffill')
exclClose = excl['Close'].tolist()
exclCloseApr = excl['2019-04']['Close'].tolist()
```

# FREN

```python
fren = fren.sort_index()
fren = fren.resample('D').sum()
fren.iloc[:,:5] = fren.iloc[:,:5].replace(0, np.NaN)
fren = fren.fillna(method='ffill')
frenClose = fren['Close'].tolist()
frenCloseApr = fren['2019-04']['Close'].tolist()
```

# ISAT

```python
isat = isat.sort_index()
isat = isat.resample('D').sum()
isat.iloc[:,:5] = isat.iloc[:,:5].replace(0, np.NaN)
isat = isat.fillna(method='ffill')
isatClose = isat['Close'].tolist()
isatCloseApr = isat['2019-04']['Close'].tolist()
```

# TLKM

```python
tlkm = tlkm.sort_index()
tlkm = tlkm.resample('D').sum()
tlkm.iloc[:,:5] = tlkm.iloc[:,:5].replace(0, np.NaN)
tlkm = tlkm.fillna(method='ffill')
tlkmClose = tlkm['Close'].tolist()
tlkmCloseApr = tlkm['2019-04']['Close'].tolist()
```

```python
date = excl.index.tolist()
plt.plot(date,exclClose,label='PT XL AXIATA')
plt.plot(date,frenClose,label='Smartfren')
plt.plot(date,isatClose,label='Indosat')
plt.plot(date,tlkmClose,label='Telkom')
plt.suptitle('Harga Historis Provider Telco Indonesia')
plt.xlabel('Tanggal')
plt.xticks(rotation=45)
plt.ylabel('Rupiah (IDR)')
plt.legend(bbox_to_anchor=(0.,1.,1.,0.1),ncol=4,mode='expand')
plt.grid(True)
plt.show()
```

```python
dateApr = excl['2019-04'].index.tolist()
plt.plot(dateApr,exclCloseApr)
plt.plot(dateApr,frenCloseApr)
plt.plot(dateApr,isatCloseApr)
plt.plot(dateApr,tlkmCloseApr)
plt.suptitle('Harga Historis Provider Telco Indonesia (April 2019)')
plt.xlabel('Tanggal')
plt.xticks(rotation=45)
plt.ylabel('Rupiah (IDR)')
plt.legend(['PT XL AXIATA','Smartfren','Indosat','Telkom'],bbox_to_anchor=(0.,1.,1.,0.1),ncol=4,mode='expand')
plt.grid(True)
plt.show()
```
